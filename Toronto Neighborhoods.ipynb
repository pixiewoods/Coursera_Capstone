### Use BeautifulSoup to scrape Toronto postal code data from https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M <br>
_Generates an index error when parsing HTML but all data is consumed and stored in dataframe_

```python
# Imports
import numpy as np                               # library to handle data in a vectorized manner
import pandas as pd                              # library for data analsysis
pd.set_option('display.max_columns', None) 
pd.set_option('display.max_rows', None)

# Imports for web scraping
import requests
import urllib.request
import time
from bs4 import BeautifulSoup

print('Libraries imported.')
```

```python
# Create new pandas dataframe
column_names = ['PostCode', 'Borough', 'Neighborhood']
toronto_hoods = pd.DataFrame(columns=column_names)
# set url to scrape from
wikiurl = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
response = requests.get(wikiurl)
# response   # 200 means good response
# Parse HTML response with beautifulsoup
soup = BeautifulSoup(response.text, "html.parser")
# Download all TR tags and take TD values
for i in range(1,len(soup.findAll('tr'))-1):
    one_tr_tag = soup.findAll('tr') [i]
    postcode = one_tr_tag.contents[1].text
    borough = one_tr_tag.contents[3].text
    neighborhood = one_tr_tag.contents[5].text.rstrip('\n')
    toronto_hoods = toronto_hoods.append({'PostCode': postcode,
                                          'Borough': borough,
                                          'Neighborhood': neighborhood}, ignore_index=True)
    time.sleep(1)
```

### Make a copy of scraped data to avoid having to go back to website.<br>
### Clean up data (dropping rows, updating missing neighborhoods) and stringing neighborhoods together.

```python
# Keep a copy of dataframe to avoid rescraping site each time I need to adjust dataframe
working_copy = toronto_hoods.copy()
working_copy
```

```python
# Remove rows without Borough ("Not assigned" )
indexBoroughs = working_copy[working_copy['Borough']=="Not assigned"].index
working_copy.drop(indexBoroughs, inplace=True)

# Update 'Not assigned' Neighborhood with value of its Borough
working_copy['Neighborhood'] = np.where(working_copy['Neighborhood'] == 'Not assigned', working_copy['Borough'], working_copy['Neighborhood'])
working_copy
```

```python
# Create dataframe with unique postcode / borough. Create neighborhood column without values.
postcode_lookup = working_copy.copy()
postcode_lookup = postcode_lookup.drop('Neighborhood', axis=1)
postcode_lookup = postcode_lookup.drop_duplicates()
postcode_lookup = postcode_lookup.assign(Neighborhood='')
postcode_lookup = postcode_lookup.reset_index(drop=True)
postcode_lookup.astype({'Neighborhood': 'object'}).dtypes
postcode_lookup
```

```python
# Concatenate Neighborhoods as single string for a PostCode

postcode_lookup.set_index('PostCode', inplace=True)

def combine_neighborhoods (post_code, borough_code, neighborhood_name):
    # print("post: " + postcode + " borough: " + boroughcode + " neighborhood: " + neighborhoodname)
    postcode = post_code
    boroughcode = borough_code
    neighborhoodname = neighborhood_name
    strNeighborhoodname = postcode_lookup.loc[postcode, "Neighborhood"]
    if len(strNeighborhoodname)>0:
        strNeighborhoodname = strNeighborhoodname + ", " + neighborhoodname
    else:
        strNeighborhoodname = neighborhoodname
        
    postcode_lookup.at[postcode, "Neighborhood"] = strNeighborhoodname
           
for row in working_copy.itertuples():
    combine_neighborhoods(row.PostCode, row.Borough, row.Neighborhood)

postcode_lookup = postcode_lookup.reset_index(drop=False)
postcode_lookup    

    
```

### More exploration with single Borough.  Create function to get category type of the venue.

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

### Clean the JSON and create dataframe.

```python
venues = results['response']['groups'][0]['items']
nearby_venues = json_normalize(venues) 
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]
# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)
# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]
nearby_venues.head()
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID,
            CLIENT_SECRET,
            VERSION,
            lat,
            lng,
            radius,
            LIMIT)
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        # return only relevant information for each nearby venue
        venues_list.append([(
            name,
            lat,
            lng,
            v['venue']['name'],
            v['venue']['location']['lat'],
            v['venue']['location']['lng'],
            v['venue']['categories'][0]['name']) for v in results])
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood',
        'Neighborhood Latitude',
        'Neighborhood Longitude',
        'Venue',
        'Venue Latitude',
        'Venue Longitude',
        'Venue Category']
    return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(names=postcode_slice['Neighborhood'],
    latitudes=postcode_slice['Latitude'],
    longitudes=postcode_slice['Longitude'])
```

### Analyze each Neighborhood

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="",prefix_sep="")
toronto_onehot.head()
# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood']
toronto_onehot.head()
# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]
toronto_onehot.head()
```

### Group rows by neighborhood and by taking the mean of the frequency of occurrence of each category

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
```

## Print each neighborhood (or multiple neighborhoods in the borough) along with the top 5 most common venues and create dataframe

```python
num_top_venues = 5

for hood in toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
    
# sort venues in descending order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
# create dataframe and display top 10 venues for each neighborhood

num_top_venues = 10
indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

## Cluster Neighborhoods

```python
# Run K-Means to cluster neighborhoods into 5 clusters

# set number of clusters
kclusters = 5
toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)
# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)
# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# Create dataframe that includes the cluster plus top 10 venues for each neighborhood

# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
toronto_merged = postcode_slice
# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')
toronto_merged.head()
```

## Create Folium map to visualize clusters

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
